##1 задание.
*Данны следующие значения истинных и предсказанных значений целевой переменной:

y_true = [1, -2, 3, 3]

y_pred = [0, -3, 2, 1]

Вычислить $MAPE, R^{2}, MAE, MSE$

```python
import numpy as np
y_true = np.array([1, -2, 3, 3])
y_pred = np.array([0, -3, 2, 1])
```

```python
#Вичислим сначала среднеквадратичную ошибку MSE
#Вручную и через sklearn.metrics
from sklearn.metrics import mean_squared_error

mse = np.mean((y_true - y_pred) ** 2)

sk_mse = mean_squared_error(y_true, y_pred)

print('By hand:', mse)
print('Auto:', sk_mse)
```

```python
#Вычислим абсолютную ошибку MAE
from sklearn.metrics import mean_absolute_error 

mae = np.mean(abs(y_true - y_pred))

sk_mae = mean_absolute_error(y_true, y_pred)
print('By hand', mae)
print('Auto', sk_mae)
```

```python
#Вычислим процентную среднюю абсолютную ошибку MAPE
from sklearn.metrics import mean_absolute_percentage_error

mape = np.mean(abs((y_true - y_pred) / y_true))

sk_mape = mean_absolute_percentage_error(y_true, y_pred)
print('By hand:', mape)
print('Auto:', sk_mape)
```

```python
#Вычислим коэффициент детерминации R^2
from sklearn.metrics import r2_score

upper = np.sum((y_true - y_pred) ** 2)
mean = y_true.mean()
lower = np.sum((y_true - mean) ** 2)

r2 = 1 - upper / lower

sk_r2 = r2_score(y_true, y_pred)

print('By hand:', r2)
print('Auto:', sk_r2)
```
