```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Courses /Старт ИИ/14.Линейная регрессия/penguins.csv')
df
```

```python
df['species'].value_counts()
```

```python
def pengton(a):
  if a == 'Adelie':
    return 0
  elif a == 'Chinstrap':
    return 1
  elif a == 'Gentoo':
    return 2
df['species'] = df['species'].apply(pengton)
```

```python
df['species'].value_counts()
```

```python
df['island'].value_counts()
```

```python
def isltona(a):
  if a == 'Biscoe':
    return 0
  elif a == 'Dream':
    return 1
  elif a == 'Torgersen':
    return 2
df['island'] = df['island'].apply(isltona)
df['island'].value_counts()
```

```python
df['sex'].value_counts()
```

```python
def setonu(a):
  if a == 'male':
    return 1
  elif a == 'female':
    return 0
df['sex'] = df['sex'].apply(setonu)
df['sex'].value_counts()
```
